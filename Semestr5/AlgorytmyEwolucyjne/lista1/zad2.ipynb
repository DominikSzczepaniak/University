```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def initial_probability_vector(d):
    return np.full(d, 0.5)

def binary_random(p):
    return np.where(np.random.rand(len(p)) < p, 1, 0)

def random_individual(p):
    return binary_random(p)

def random_population(p, N):
    return np.array([random_individual(p) for _ in range(N)])

def population_evaluation(P, F):
    return np.array([F(individual) for individual in P])

def best_individual(P, fitness):
    index = np.argmax(fitness)
    return P[index]
```

```python
def pbil(F, d, N, theta1, theta2, theta3, iterations):
    p = initial_probability_vector(d)
    P = random_population(p, N)
    fitness = population_evaluation(P, F)
    
    for _ in range(iterations):
        xi = best_individual(P, fitness)

        p = p * (1 - theta1) + xi * theta1

        mutation_indices = np.random.rand(d) < theta2
        mutation_values = binary_random(np.full(d, 0.5))

        p[mutation_indices] = p[mutation_indices] * (1 - theta3) + mutation_values[mutation_indices] * theta3

        P = random_population(p, N)
        fitness = population_evaluation(P, F)
    xi = best_individual(P, fitness)
    return xi, F(xi)

def pbil_charts(F, d, N, theta1, theta2, theta3, iterations):
    p = initial_probability_vector(d)
    P = random_population(p, N)
    fitness = population_evaluation(P, F)
    best_fitness_history = []
    average_fitness_history = []
    
    for _ in range(iterations):
        xi = best_individual(P, fitness)
        best_fitness = F(xi)
        best_fitness_history.append(best_fitness)
        average_fitness_history.append(np.mean(fitness))
        
        p = p * (1 - theta1) + xi * theta1
        
        mutation_indices = np.random.rand(d) < theta2
        mutation_values = binary_random(np.full(d, 0.5))
        
        p[mutation_indices] = p[mutation_indices] * (1 - theta3) + mutation_values[mutation_indices] * theta3
        
        P = random_population(p, N)
        fitness = population_evaluation(P, F)
    
    xi = best_individual(P, fitness)
    best_fitness = F(xi)
    best_fitness_history.append(best_fitness)
    average_fitness_history.append(np.mean(fitness))
    
    return xi, best_fitness, best_fitness_history, average_fitness_history
```

```python
def onemax(individual):
    return np.sum(individual)

def deceptive_onemax(individual):
    n = len(individual)
    u = np.sum(individual)
    if u == n:
        return n
    else:
        return n - 1 - u

def k_deceptive_onemax(individual, k):
    total = 0
    for i in range(0, len(individual), k):
        block = individual[i:i+k]
        u = np.sum(block)
        if u == k:
            total += k
        else:
            total += k - 1 - u
    return total

def k_deceptive_onemax_wrapper(individual):
    return k_deceptive_onemax(individual, k=5)
```

```python
d_values = [50, 100, 200]  
theta1_values = [0.1, 0.3, 0.5]
theta2_values = [0.02, 0.05, 0.1]
theta3_values = [0.1, 0.3, 0.5]
iterations = 100
N = 1000

for d in d_values:
    if d % 5 != 0:
        continue
    print(f"\nDługość chromosomu: {d}")
    
    plt.figure(figsize=(10, 5))
    plt.title(f'OneMax - Porównanie różnych parametrów, d={d}')
    plt.xlabel('Iteracja')
    plt.ylabel('Fitness')

    for theta1 in theta1_values:
        for theta2 in theta2_values:
            for theta3 in theta3_values:
                print(f"\nParametry: θ1={theta1}, θ2={theta2}, θ3={theta3}")
                
                best_solution, best_fitness, best_fitness_history, avg_fitness_history = pbil_charts(onemax, d, N, theta1, theta2, theta3, iterations)
                print("Najlepsza wartość:", best_fitness)

                plt.plot(best_fitness_history, label=f'θ1={theta1}, θ2={theta2}, θ3={theta3}')
    
        plt.legend()
        plt.show()
```
