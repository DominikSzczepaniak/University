## Objective function

```python
def tsp_objective_function(p, n, A):
    s = 0.0
    for i in range(n):
        s += A[p[i-1], p[i]]
    return s
```

## Simulated Annealing

```python
def random_neighbor(p, radius):
    q = p.copy()
    for r in range(radius):
        i, j = np.random.choice(n, 2, replace=False)
        q[i], q[j] = q[j], q[i]
    return q
```

```python
T = 500000
radius = 1
alpha = 1.0

t0 = time.time()

p = np.random.permutation(n)
p_cost = tsp_objective_function(p)
costs_simulated = np.zeros(T)
for t in range(T):
    q = random_neighbor(p, radius)
    q_cost = tsp_objective_function(q)
    if(q_cost < p_cost):
        p, p_cost = q, q_cost
    elif(np.random.rand() < np.exp(- alpha * (q_cost - p_cost) * t/T)):
        p, p_cost = q, q_cost
    costs_simulated[t] = p_cost

print(time.time() - t0, costs_simulated.min())
```

## SGA-PMX

```python
def PMX(ind1, ind2, separator_no=2): 
    new_ind1, new_ind2 = ind1.copy(), ind2.copy()
    idxs = sorted(np.random.choice(len(ind1), separator_no, replace=False))
    
    group = np.random.choice(separator_no-1)
    start, end = idxs[group], idxs[group+1]
    
    tmp = ind1[start:end].copy()
    ind1[start:end] = ind2[start:end]
    ind2[start:end] = tmp
    
    for i in range(len(ind1)):
        if start <= i < end:
            continue
            
        while ind1[i] in ind1[start:end]:
            idx_of_elem = np.nonzero(ind1[start:end] == ind1[i])[0][0]
            ind1[i] = ind2[start+idx_of_elem]
        
        while ind2[i] in ind2[start:end]:
            idx_of_elem = np.nonzero(ind2[start:end] == ind2[i])[0][0]
            ind2[i] = ind1[start+idx_of_elem]

    return ind1, ind2
```

```python
def reverse_sequence_mutation(p):
    a = np.random.choice(len(p), 2, False)
    i, j = a.min(), a.max()
    q = p.copy()
    q[i:j+1] = q[i:j+1][::-1]
    return q
```

```python
def random_transposition_mutation(p):
    a = np.random.choice(len(p), 2, False)
    i, j = a.min(), a.max()
    q = p.copy()
    q[i], q[j] = q[j], q[i]
    return q
```

```python
def SGA(mutation_function, n, A, coords, number_of_iterations=250):
    population_size = 500
    chromosome_length = n
    number_of_offspring = population_size
    crossover_probability = 0.95
    mutation_probability = 0.25

    time0 = time.time()

    best_objective_value = np.inf
    best_chromosome = np.zeros((1, chromosome_length))
    costs_SGA = np.zeros(number_of_iterations)

    # generating an initial population
    current_population = np.zeros((population_size, chromosome_length), dtype=np.int64)
    for i in range(population_size):
        current_population[i, :] = np.random.permutation(chromosome_length)

    # evaluating the objective function on the current population
    objective_values = np.zeros(population_size)
    for i in range(population_size):
        objective_values[i] = tsp_objective_function(current_population[i, :], n, A)

    for t in range(number_of_iterations):
        # selecting the parent indices by the roulette wheel method
        fitness_values = objective_values.max() - objective_values
        if fitness_values.sum() > 0:
            fitness_values = fitness_values / fitness_values.sum()
        else:
            fitness_values = np.ones(population_size) / population_size
        parent_indices = np.random.choice(population_size, number_of_offspring, True, fitness_values).astype(np.int64)
        # creating the children population
        children_population = np.zeros((number_of_offspring, chromosome_length), dtype=np.int64)
        for i in range(int(number_of_offspring/2)):
            if np.random.random() < crossover_probability:
                children_population[2*i, :], children_population[2*i+1, :] = PMX(current_population[parent_indices[2*i], :].copy(), current_population[parent_indices[2*i+1], :].copy())
            else:
                children_population[2*i, :], children_population[2*i+1, :] = current_population[parent_indices[2*i], :].copy(), current_population[parent_indices[2*i+1]].copy()

        if np.mod(number_of_offspring, 2) == 1:
            children_population[-1, :] = current_population[parent_indices[-1], :]

        # mutating the children population
        for i in range(number_of_offspring):
            if np.random.random() < mutation_probability:
                children_population[i, :] = mutation_function(children_population[i, :])

        # evaluating the objective function on the children population
        children_objective_values = np.zeros(number_of_offspring)
        for i in range(number_of_offspring):
            children_objective_values[i] = tsp_objective_function(children_population[i, :], n, A)

        # replacing the current population by (Mu + Lambda) Replacement
        objective_values = np.hstack([objective_values, children_objective_values])
        current_population = np.vstack([current_population, children_population])

        I = np.argsort(objective_values)
        current_population = current_population[I[:population_size], :]
        objective_values = objective_values[I[:population_size]]

        # recording some statistics
        if best_objective_value < objective_values[0]:
            best_objective_value = objective_values[0]
            best_chromosome = current_population[0, :]
        costs_SGA[t] = objective_values.mean()

        # print('%3d %14.8f %12.8f %12.8f %12.8f %12.8f' % (t, time.time() - time0, objective_values.min(), objective_values.mean(), objective_values.max(), objective_values.std()))
    return costs_SGA
```

```python
%%capture
def run_SGA_multiple(number_of_runs, mutation_function, n, A, coords, iterations=250):
    #run SGA multiple times and mean the costs_SGA
    costs_SGA_multiple = np.zeros((number_of_runs, iterations))
    for i in range(number_of_runs):
        costs_SGA_multiple[i] = SGA(mutation_function, n, A, coords, iterations)
    final_costs = np.mean(costs_SGA_multiple, axis=0)
    return final_costs
```

```python
#https://github.com/pdrozdowski/TSPLib.Net/tree/master/TSPLIB95/tsp

def read_tsp_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    n = 0
    coors = []
    dists = []
    upper_row_data = []
    is_in_edge_weight_section = False
    is_in_display_data_section = False
    
    for line in lines:
        line = line.strip()
        
        # Omiń nagłówkowe informacje
        if line.startswith("NAME") or line.startswith("TYPE") or line.startswith("COMMENT"):
            continue
        if line.startswith("DIMENSION"):
            n = int(line.split(":")[1].strip())
            dists = np.zeros((n, n))  
        if line.startswith("EDGE_WEIGHT_SECTION"):
            is_in_edge_weight_section = True
            continue
        if line.startswith("DISPLAY_DATA_SECTION"):
            is_in_edge_weight_section = False
            is_in_display_data_section = True
            continue
        if line.startswith("EOF"):
            break
        
        if is_in_edge_weight_section:
            upper_row_data.extend(map(int, line.split()))
        
        if is_in_display_data_section:
            parts = line.split()
            x, y = float(parts[1]), float(parts[2])
            coors.append((x, y))
    
    index = 0
    for i in range(n):
        for j in range(i + 1, n):
            dists[i, j] = upper_row_data[index]
            dists[j, i] = upper_row_data[index]
            index += 1

    coors = np.array(coors)
    return n, coors, dists
```

```python
def solve_problem(problem_name):
    n, coords, A = read_tsp_data(f"{problem_name}.tsp")
    costs_SGA1 = run_SGA_multiple(2,random_transposition_mutation, n, A, coords)
    costs_SGA2 = run_SGA_multiple(2,reverse_sequence_mutation, n, A, coords)
    
    plt.title(f'{problem_name} - SGA, best cost for transposition: %f, best cost for reverse sequence mutation: %f' % (costs_SGA1.min(), costs_SGA2.min()))
    plt.plot(costs_SGA1)
    plt.plot(costs_SGA2)
    plt.legend(['Transposition', 'Reverse sequence mutation'])
    plt.show()
```

```python
def read_tsp_euc2d_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    n = 0
    coors = []
    is_in_node_coord_section = False
    
    for line in lines:
        line = line.strip()
        
        if line.startswith("NAME") or line.startswith("TYPE") or line.startswith("COMMENT"):
            continue
        if line.startswith("DIMENSION"):
            n = int(line.split(":")[1].strip())
            dists = np.zeros((n, n)) 
        if line.startswith("NODE_COORD_SECTION"):
            is_in_node_coord_section = True
            continue
        if line.startswith("EOF"):
            break
        
        if is_in_node_coord_section:
            parts = line.split()
            x, y = float(parts[1]), float(parts[2])
            coors.append((x, y))
    
    coors = np.array(coors)
    
    for i in range(n):
        for j in range(i + 1, n):
            dist = np.sqrt((coors[i][0] - coors[j][0]) ** 2 + (coors[i][1] - coors[j][1]) ** 2)
            dists[i, j] = dist
            dists[j, i] = dist

    return n, coors, dists
```

```python
def solve_euc2d_problem(problem_name, iterations=250):
    n, coords, A = read_tsp_euc2d_data(f"{problem_name}.tsp")
    costs_SGA1 = run_SGA_multiple(2,random_transposition_mutation, n, A, coords, iterations)
    costs_SGA2 = run_SGA_multiple(2,reverse_sequence_mutation, n, A, coords, iterations)
    
    plt.title(f'{problem_name} - SGA, best cost for transposition: %f, best cost for reverse sequence mutation: %f' % (costs_SGA1.min(), costs_SGA2.min()))
    plt.plot(costs_SGA1)
    plt.plot(costs_SGA2)
    plt.legend(['Transposition', 'Reverse sequence mutation'])
    plt.show()
```
